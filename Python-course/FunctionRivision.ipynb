```python
def fun11(*args):
    d = {}
    for i in args:
        if type(i) in d:
            d[type(i)].append(i)
        else:
            d[type(i)] = [i]
    return d
```

```python
fun11([2,3,4,4],"ask",3,(2,3,4), True,3,4,5)
```

```python
def fun12(*args, **kwargs):
    r = []
    for i in args:
        if type(i) == list:
            r.extend(i)
    for i in kwargs.values():
        if type(i) == list:
            r.extend(i)
    return r
```

```python
fun12([1,2,3,4,5], [22,33,44],2,3,4,a= 20,b= 34,c = [2,3,4,55555])
```
