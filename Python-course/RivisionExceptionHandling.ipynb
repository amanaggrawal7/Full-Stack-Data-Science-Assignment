```python
def askforint():
    while True:
        try:
            a = int(input("Enter an integer"))
        except FileNotFoundError as e:
            print("This is error message", e)
        except IOError as e:
            print(e)
        except ValueError as e:
            print(e)
        else:
            print("person has entered a corrct value")
            break
        finally:
            print("Close the issue")
```
