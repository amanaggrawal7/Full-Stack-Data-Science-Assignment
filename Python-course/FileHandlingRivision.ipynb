```python
import os
import shutil

src = r"D:\FileHandlingRivision\Rivision\cw"
dst = r"D:\FileHandlingRivision\Rivision\hw"

os.makedirs(dst, exist_ok=True)

for item in os.listdir(src):
    if item == "newfile":   # skip this file/folder
        continue

    shutil.move(
        os.path.join(src, item),
        os.path.join(dst, item)
    )
```
