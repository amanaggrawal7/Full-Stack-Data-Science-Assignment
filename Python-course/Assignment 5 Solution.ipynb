```python
#We have to find Lcm Of two number
```

```python
def FindLcm(a,b):
    if a>b:
        s = a
    else:
        s = b
    while True:
        if s % a == 0 and s % b == 0:
            m = s
            break
        else:
            s +=1
    return m
    
        
```

```python
FindLcm(2,3)
```

```python
FindLcm(12,24)
```

```python
FindLcm(5,9)
```

```python
# Write a program to find HCF
```

```python
def FindHcf(a,b):
    if a>b:
        s = b
    else:
        s = a
    for i in range(1,s+1):
        if (a%i == 0) and (b%i==0):
            hcf = i
    print("Hcf of {} and {} is {}".format(a,b,hcf))
```

```python
FindHcf(1234,2344)
```

```python
def decimaltoOther():
    num = int(input("Enter a number :"))
    print(f"Binary Number is {bin(num)}")
    print(f"octal number is {oct(num)}")
    print(f"hexadecimal number is {hex(num)}")
```

```python
def charToAscii():
    char = input("Enter a char value")
    if len(char)>1:
        print("Enter only single char")
    else:
        print(f"the Ascii value of {char} is {ord(char)}")
```
