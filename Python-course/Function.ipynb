#### create a function to take any number of data aand try to create a list of sepearte data based on data type and return multiple result using args

```python
def test19(*args):
    lstr = []
    lint = []
    lbool = []
    llist = []
    lall = []
    
    for i in args:
        if type(i) == str:
            lstr.append(i)
        elif type(i) == int:
            lint.append(i)
        elif type(i) == bool:
            lbool.append(i)
        elif type(i) == list:
            llist.append(i)
        else:
            lall.append(i)
    return lstr,lint,lbool,llist,lall
    
```

#### create a function which will be able to use *args and *kwagrs and it will be able to do all the list value concatenation and return a list

```python
test19("aman",12,"kumar",[2,3,4,5],"halo",[2,3,45,6,6], (2,3,4),3+2j)
```

```python
def test19(*args, **kwargs):
    result = []

    for i in args:
        if type(i) == list:
            result.extend(i)
    for v in kwargs.values():
        if type(v) == list:
            result.extend(v)
    return result
```

```python
test19([2,3,4],[23,43,3],a = [2,3,4,5])
```

```python
def concat_all_lists(*args, **kwargs):
    result = []

    # process positional arguments
    for item in args:
        if type(item) == list:
            result.extend(item)

    # process keyword arguments
    for value in kwargs.values():
        if type(value) == list:
            result.extend(value)

    return result
```

```python
concat_all_lists([2,3,4],[23,43,3],a = [2,3,4,5])
```
