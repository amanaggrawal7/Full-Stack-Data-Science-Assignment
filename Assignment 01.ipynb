```python
a = int(input("Enter a 1st number"))
b = int(input("Enter a second number"))
print("The addition of number is ", a+b)
if (b ==0):
    print("Division not possible")
else:
    print("Division is ", a/b)
```

3.Write a Python program to find the area of a triangle ?

```python
base = int(input("Enter a base"))
height = int(input("Enter a height"))
print("Area of triangle is ", 0.5*base*height)
```

4.Write a Python program to swap two variables ?

```python
a = int(input("Enter a 1st number a = "))
b = int(input("Enter a second number b = "))
temp = a
a = b
b = temp
print("Swap number is a = ", a)
print("Swap number is b = ", b)
```

5.Write a Python program to generate a random number ?

```python
from random import randit
def generateRandomNumber(start = 0, end = 100000):
    print("Random number -> ", randit(start, end))

generateRandomNumber() # without arguments

generateRandomNumber(0, 1000) #With arguments
```
